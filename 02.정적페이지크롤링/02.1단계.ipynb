```python
""" 
카테고리, 상품명, 상세피이지 링크, 가격 - 4가지 정보 추출
"""

import requests
from bs4 import BeautifulSoup
response = requests.get("https://startcoding.pythonanywhere.com/basic")
html = response.text
soup = BeautifulSoup(html, 'html.parser')
```

```python
# 1. 카테고리
category = soup.select_one(".product-category").text
print(category)
```

```python

# 2. 상품명
product_name = soup.select_one(".product-name>a").text
print(product_name)
```

```python
# 상세페이지 링크
product_link = soup.select_one(".product-name>a").attrs['href']
print(product_link)
```

```python
# 가격
soup.select_one(".product-price").text # 앞 뒤로 escape 공백 문자가 붙음
```

```python
# strip() : escape 공백 문자 앞뒤로 제거
soup.select_one(".product-price").text.strip() # escape 공백 문자 앞뒤로 제거 : strip()
```

```python
 # 변수에 할당할 경우 escape 공백 문자가 출력되지 않음
product_price = soup.select_one(".product-price").text
print(product_price)
```

```python
# replace(): 문자열 치환
soup.select_one(".product-price").text.strip().replace(',', '').replace('원', '')
```

```python
# 완성 코드
import requests
from bs4 import BeautifulSoup
response = requests.get("https://startcoding.pythonanywhere.com/basic")
html = response.text
soup = BeautifulSoup(html, 'html.parser')

category = soup.select_one(".product-category").text
name = soup.select_one(".product-name>a").text
link = soup.select_one(".product-name>a").attrs['href']
price = soup.select_one(".product-price").text.strip().replace(',', '').replace('원', '')
print(category, name, link, price)
```
